```python
iptv_orig = iptv_orig[['user_id', 'content_code','item_code','play_start_time','play_duration','duration']]
```

```python
iptv_orig['play_start_time'] = iptv_orig['play_start_time'].apply(pd.to_datetime)
iptv_orig['timestamp']=iptv_orig['play_start_time'].values.astype(int)
#iptv_orig['play_time'] = pd.to_datetime(iptv_orig['play_ts'], unit='ns')
```

```python
interactions=iptv_orig.groupby(["user_id","item_code"]).agg({"timestamp": min,"content_code": np.count_nonzero, "play_duration": np.sum, "duration": np.sum}).reset_index()
interactions.head(5)
```

```python
interactions.info()
```

```python
interactions['PLAY_PCT']=np.round(interactions['play_duration']/interactions['duration'],2)
interactions.rename(columns = {'user_id':'USER_ID', 'item_code':'ITEM_ID', 'content_code':'content_count', 
                              'timestamp':'TIMESTAMP'}, inplace = True) 
interactions.head(5)
```

```python

df = interactions.groupby("ITEM_ID").agg({"USER_ID": np.count_nonzero, "PLAY_PCT": np.average})
df.info()
#df.sort_values('USER_ID',ascending=False)
#df.sort_values('ITEM_ID',ascending=False)
#df.head(5)
df[df["USER_ID"]>5].count()

df = interactions.groupby("USER_ID").agg({"ITEM_ID": np.count_nonzero,"PLAY_PCT": np.average})
df.info()
df[df["ITEM_ID"]>3].count()
```

```python
click_df=interactions.copy()
click_df=click_df[click_df['PLAY_PCT']<0.1]
click_df['EVENT_TYPE']='click'
click_df.head(5)
```

```python
watched_df=interactions.copy()
watched_df=watched_df[(watched_df['PLAY_PCT']>=0.1) & (watched_df['PLAY_PCT']<0.6)]
watched_df['EVENT_TYPE']='watch'
watched_df.head()
```

```python
loved_df=interactions.copy()
loved_df=loved_df[(loved_df['PLAY_PCT']>=0.6)]
loved_df['EVENT_TYPE']='finish'
loved_df.head()
```

```python

actions_df = click_df.copy()
actions_df = actions_df.append(watched_df)
actions_df = actions_df.append(loved_df)
actions_df.sort_values("TIMESTAMP", axis = 0, ascending = True, 
                 inplace = True, na_position ='last') 
actions_df.drop('duration',axis=1, inplace=True)
actions_df.drop('PLAY_PCT',axis=1, inplace=True)
actions_df.head()
```

```python
interactions_filename = "interactions.csv"
actions_df.to_csv(interactions_filename, index=False, float_format='%.0f')
```

```python
content_orig['region']=content_orig['region'].str.replace('百视通','NaN')\
                        .replace('2019','NaN').replace('2017','NaN').replace('无','NaN')\
                        .str.replace('/','').str.replace('"','').str.replace(']','').str.replace('[','')\
                        .str.replace(',',' ').str.replace('、',' ')
content_orig['region'].unique()
```

```python
content_orig['region']=content_orig['region'].str.replace('zh-cn','中国')\
                            .str.replace('中国大陆','中国')\
                            .str.replace('中国台湾','台湾')\
                            .str.replace('中国香港','香港')\
                            .str.replace('中国内地','中国')
content_orig['region']=content_orig['region'].str.split().str.join('|')
content_orig['region'].unique()
```

```python
content_orig['region']=content_orig['region'].str.replace('大陆','中国')\
                            .str.replace('内地','中国').str.replace('英语','英国')\
                            .str.replace('其他','NaN')
content_orig['region']=content_orig['region'].str.replace('台湾\|台湾','台湾').str.replace('中国\|中国','中国')
content_orig['region'].unique()
```

```python
item_df=content_orig.groupby(["item_code","region","issue_year","content_type","genre","actor_display","writer_display","tags"]).agg({"duration": np.sum}).reset_index()
item_df.head(5)
item_df.info()
```
