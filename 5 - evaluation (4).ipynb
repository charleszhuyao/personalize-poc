```python
df_valid = pd.read_csv('./poc_data/inter_for_valid.csv')
```

```python
df_valid.head()
```

```python
res_dict = {}
```

```python
%%time

for user_id in df_valid['user_id'].values.tolist():
        get_recommendations_response = personalize_runtime.get_recommendations(
        campaignArn = 'arn:aws:personalize:us-east-1:662566784674:campaign/dtv-v1',
        userId = user_id,
        numResults = 16)
        
        res_dict[user_id]={
            'item_id' : df_valid[df_valid['user_id']==user_id].iloc[0]['item_code'],
            'rec_list' : get_recommendations_response['itemList']
        }
```

```python
res_dict
```

```python
print('the real:\n%40s %s\nthe rec list:' % (res_dict[user_id]['item_id'], get_item_name(res_dict[user_id]['item_id'])))
for item in res_dict[user_id]['rec_list']:
    print('%40s %s' % (item['itemId'], get_item_name(item['itemId'])))
```

```python
hit_list=[]
miss_list=[]

for user_id, value in res_dict.items():
    desired_item = value['item_id']
    rec_set = set(map(lambda x:x['itemId'],value['rec_list']))
    
    if desired_item in rec_set:
        hit_list.append(user_id)
    else:
        miss_list.append(user_id)
```
